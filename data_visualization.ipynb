```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
gas90_95 = gas[:6]
gas90_95
```

```python
plt.figure(figsize=(10 ,6))
plt.plot(gas.Year ,gas.Australia ,  'b.-' , label = 'Austrlia')
plt.plot(gas.Year ,gas.Canada , 'r.-' , label = 'Canada')
plt.xticks(gas.Year.values)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10 ,6))
plt.plot(gas.Year ,gas.USA ,    label = 'USA')
plt.plot(gas.Year ,gas.Canada ,  label = 'Canada')
plt.xticks(gas.Year.values)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10 ,6))
plt.plot(gas.Year ,gas.USA ,    label = 'USA')
plt.plot(gas.Year ,gas.Canada ,  label = 'Canada')
plt.plot(gas.Year ,gas.France ,  label = 'France')
plt.plot(gas.Year ,gas.UK ,  label = 'UK' , linewidth =3)
plt.plot(gas.Year ,gas['South Korea'] ,  label = 'South Korea')
plt.xticks(gas.Year.values)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Gas prices in USA and Canada")

countries = ["USA", "Canada", "Italy", "UK" , "France"]

for i in gas:
    if i in countries:
        plt.plot(gas.Year, gas[i] , label = i)
        
plt.plot(gas.Year , gas["South Korea"] , label= "South Korea", linewidth = 3)

plt.xlabel("Year")
plt.ylabel("US Dollars")
plt.legend()
plt.xticks(gas.Year[::2])
plt.show()
```

```python
plt.figure(figsize=(8,5))
for country in gas.iloc[: ,1:] : 
    plt.plot(gas.Year , gas[country] ,label = country )
plt.xlabel("Year")
plt.ylabel("US Dollars")  
plt.legend()
plt.xticks(gas.Year[::2])
plt.show()     
```
