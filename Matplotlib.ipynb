```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
```

```python
x = [1 , 2, 3 , 4]
y = [2 , 4 , 8 , 12]
plt.scatter(x ,y)
```

```python
plt.plot(x , y)
```
